```python
"""
PAN 15 TWEET XML TO CSV
"""

import os
import pandas as pd
import xmltodict

dir_path = "/Users/alexlain/github/author-profiling-project/data/pan-15/train"

df_master = pd.DataFrame()

for file in os.listdir(dir_path):
    if file.endswith(".xml"):
        with open(dir_path + "/" + file, 'r') as f:
            xml = f.read()
    
        xml_dict = xmltodict.parse(xml)
        xml_dict

        df = pd.json_normalize(xml_dict,
                       record_path=['author', 'document'],
                       meta=[['author', '@id']])

        df_training = pd.concat([df_training, df])

df_training.rename(columns={0: 'text', 'author.@id': 'author'}, inplace=True)
df_training
```

```python
df_training.to_csv(dir_path + "/" + "pan-15-train.csv")
```

```python
"""
PAN 15 TRUTH TXT TO CSV
"""

#userid:::gender:::age_group:::extroverted:::stable:::agreeable:::conscientious:::open

file_path = "/Users/alexlain/github/author-profiling-project/data/pan-15/train/truth.txt"

df_truth = pd.read_csv(file_path, sep=":::", header=None, engine='python')
df_truth.columns = ["author", "gender", "age-group", "extroverted", "stable", "agreeable", "conscientious", "open"]

df_truth
```
